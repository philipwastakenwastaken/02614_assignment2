```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns

sns.set()
```

```python
rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
rc('text', usetex=True)
```

```python
jacobi_it = np.loadtxt("data/jaco.txt", usecols=0)
gs_it = np.loadtxt("data/gs.txt", usecols=0)
jacobi_time = np.loadtxt("data/jaco.txt", usecols=1)
gs_time = np.loadtxt("data/gs.txt", usecols=1)

x = np.array(range(10, 110, 10))
```

```python

plt.plot(x, jacobi_it, '-o', label="Jacobi")
plt.plot(x, gs_it, '-v', label="Gauss-Seidel")

plt.xlabel('$N$')
plt.ylabel('Iterations')

plt.grid(True, which="major", ls="-", color='black', alpha=0.2)

plt.legend()
plt.tight_layout()
plt.savefig('img/seq_it.pdf', format='pdf', transparent=True)
plt.show()
```

```python
jacobi_grid_updates = jacobi_it / jacobi_time
gs_grid_updates = gs_it / gs_time
```

```python
plt.plot(x, jacobi_grid_updates, '-o', label="Jacobi")
plt.plot(x, gs_grid_updates, '-v', label="Gauss-Seidel")

plt.xlabel('$N$')
plt.ylabel('Iterations per second')

plt.grid(True, which="major", ls="-", color='black', alpha=0.2)

plt.legend()
plt.yscale('log')
plt.tight_layout()
plt.savefig('img/seq_it_per_second.pdf', format='pdf', transparent=True)
plt.show()
```

```python
plt.plot(x, jacobi_time, '-o', label="Jacobi")
plt.plot(x, gs_time, '-v', label="Gauss-Seidel")

plt.xlabel('$N$')
plt.ylabel('Runtime [s]')

plt.grid(True, which="major", ls="-", color='black', alpha=0.2)

plt.legend()
#plt.yscale('log')
plt.tight_layout()
plt.savefig('img/seq_runtime.pdf', format='pdf', transparent=True)
plt.show()
```
